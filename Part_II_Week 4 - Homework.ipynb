####Question 3
This problem explores the difference between internal and external habits in a very simple environment. We'll assume perfect foresight -- no Et terms -- and $\beta=1/Rf$. The consumer has utility with one-period habit formation.
$$U = \sum \beta^t u\left(C_t - \theta C_{t-1}\right)$$
or, near time $t$,

$$U = ...  \beta^{t-1} u\left(C_{t-1} - \theta C_{t-2}\right) 
 +\beta^{t} u\left(C_{t} - \theta C_{t-1} \right)
+\beta^{t+1} u\left(C_{t+1} - \theta C_{t} \right)+...$$
Now, we compute the marginal utility of consumption at time $t$. With external habit, we ignore the presence of $\theta C_t$ in utility at time $t+1$. (More formally, $\theta C_t$ represents aggregate consumption over which the investor has no control. ) Thus, the external-habit marginal utility of consumption at time $t$ is
$$MU(C_t) = \frac{\partial U(C_t)}{\partial C_t} = u'(C_t - \theta C_{t-1})$$

and the riskfree rate equation is
$$MU(C_t) = MU(C_{t+1})$$
$$u'(C_t - \theta C_{t-1}) = u'(C_{t+1} - \theta C_{t}) \; \;  (*)$$
(Remember, $\beta R^f=1$, and perfect foresight.)

Now, let's do the same thing with internal habits. In this case, the investor recognizes that consumption today will influence habits in the future. So, take the derivative of the utility function with respect to $C_t$, including all places where $C_t$ appears. You should see a "strategic" term whereby the consumer thinks about how consumption today affects habits in the future. Calculate $MU(C_t)$ and $MU(C_{t+1})$, and express the first order condition corresponding to (*).

Now, suppose the consumption process $\{C_t\}$ satisfies the external-habit first-order condition (*) in this case. Does that consumption process also satisfy the internal-habit first order condition that you just derived? 

>**Answer: Any consumption process that solves the external-habit first order conditions (*) also solves the internal-habit first order conditions. **

####Question 7
Now, find the basic asset pricing equation in this two period model. Take $\partial U)t/\partial C_t$, and then $\partial U_t/\partial C_{t+1}$. In the latter, remember you are really taking $\partial U_t/\partial C_{t+1}(s)$ in one state $s$. Writing the expectation out as a sum may help you to get that right. Then write the basic asset pricing equation, $p_t \partial U_t/ \partial C_t = E_t \left[ \partial U_t / \partial C_{t+1} x_{t+1}\right]$, i.e. the equation that generalizes $p_t C_t^{-\gamma} = \beta E_t \left[ C_{t+1}^{-\gamma} x_{t+1} \right]$

>**Answer: $p_t C_t^{-\rho}  = \beta \left[ E_t ( C_{t+1}^{1-\gamma})\right] ^ \frac{\gamma-\rho}{1-\gamma} E_t \left[ C_{t+1}^{-\gamma} x_{t+1} \right]$** 

>The first $E_t$ term comes out of the expectation. So recursive utility here only affects the interest rate. Risk premiums are driven entirely by the conventional second term. That's a more general result. The point is to separate EIS from risk aversion. Risk aversion then is pretty standard power utility stuff. This does have deeper effects in dynamic models, since today's consumption won't reveal future news (that's EIS stuff) as it did before. But fundamentally, that's why recursive utility models have to add one of the usual suspects like time varying consumption volatility (or habits?
