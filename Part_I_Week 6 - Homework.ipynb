####Question 2

The CAPM is
$E(R_i)−R^f=β_{i,m}[E(R_m)−R^f]$
You can difference the left hand variables, to produce a CAPM with excess returns defined as one risky asset return minus another rather than one risky asset return minus the risk free rate,
$E(R_i−R_j)=β_{i−j,m}[E(R_m)−R^f]$
(Covariance is a linear operator, so $β_{i,m}−β_{j,m}=β_{i−j,m}$.) Can we can do this on the right hand side too? Can we write
$E(R^e_i)=β_i,R_m−R_kE(R_m−R_k)$,
where $R_k$ is the return on a generic risky asset? Note, $β_i,R_m−R_k$ means that we run the time series regression with $R_m−R_k$ on the right hand side, so it's not trivial!



Check the correct answers, leave blank the incorrect answers. If the CAPM is correct...

>**Answer: The right hand variable can be $R_m−R_k$ if $R_k$ is mean-variance efficient, meaning $R_k=R_f+w(R_m−R_f)$ for some $w$.**

>**Answer: The right hand variable can be $R_k−R_j$ if both $R_k$ and $R_j$ are mean-variance efficient, meaning $R_k=R_f+w_k(R_m−R_f)$ for some $w_k$ and similarly for $R_j$ with $w_j\neq w_k$.**

####Question 4
When we set up the infinite-horizon CAPM and ICAPM we looked at the price of a consumption claim
$$p_{t} =E_{t}\sum_{j=1}^{\infty }\beta ^{j}\left( \frac{c_{t+j}}{c_{t}}\right) ^{-\gamma }c_{t+j}$$
$$p_{t} = E_{t}\int_{s=0}^{\infty }e^{-\delta s}\left( \frac{c_{t+s}}{c_{t}}\right) ^{-\gamma }c_{t+s}ds$$
Suppose that we get news today of good consumption in the future -- news that $c_{t+j}$ or $c_{t+s}ds$ will be higher, but there is no change in consumption today $c_t$. What effect does this have on the consumption-claim price today $p_t$? 

>**Answer: For log utility p/c is constant no matter what kind of news we get about future consumption! For $\gamma>1$, higher future consumption actually lowers the price today!
What's going on? Well, consumption shows up both as a cashflow and as a discount factor. When we get good news of future consumption, the discount rate rises, and that more than offsets the cashflow effect. As a simple intuition, with perfect foresight (no uncertainty) the j period interest rate (return on bond) is $R(j)=\frac{1}{\beta^j(\frac{c_{t+j}}{c_t})^{-\gamma}}=\frac{1}{β_j}(\frac{c_{t+j}}{c_t})^\gamma$. Thus, a higher $c_{t+j}$ raises the interest rate -- for $\gamma>1$, if people get news of good future consumption they want to eat now, and for $\gamma>>1$, they must be offered a huge interest rate to keep them from doing it. Big picture: there are always discount rate and cashflow effects! Markets are more subtle than you think, and don't jump to say they're irrational when you can't figure them out. (Ok, I'm a bit sensitive while writing this, Fama's Nobel week!)
This is an important problem when reading the news. In our macro discussion, I noticed that lower interest rates in recessions didn't necessarily have anything to do with the Fed -- real interest rates would of course fall when expected consumption growth falls.
Often, you will see a puzzle that stock markets fall on good economic news. Commenters will say "People expect the Fed to raise rates sooner on this news, and that's why stocks fell." But again, the discount rate based stock price fall will happen even in this perfectly frictionless model.
In economic parlance, there is an income and a substitution effect. For log utility income and substitution effects exactly offset. This happens also in labor economics: Overall hours of work are pretty insensitive to wages. If you are offered twice as much wage, you say, great I'll work harder -- substitute leisure to labor -- but you also feel wealthier which makes you want to work less. The two effects offset to a large extent. **

####Question 8
You are running an event study, to measure the effect of an announcement on a company's stock price, and you want to isolate the component of a company's return on a given day that corresponds to firm-specific news, not market news or general industry news. 

>**Answer: This is a case in which you don't care about mean returns at all, you're just getting rid of sources of variance. Here, you don't need traded assets, you need the best measures of true influence. Run
$$R^{ei}_t=α_i+b_irmrf_t+c_ihml_t+d_iR^{industry}_t+\epsilon^i_t$$
then form
$$R^{ei}_t−b_irmrf_t+c_ihml_t+d_iR^{industry}_t$$
which is your guess of the idiosyncratic return shock on that day.**


####Question 9
You're starting a hedge fund. Your clients demand that you have no market or value exposure and want as little volatility as possible. 

>**Answer: For this application, it's really important that the factors are traded, because you're going to hedge them out. Run $R^{ei}_t=\alpha_i+b_iSP500_t+h_iETF_t+c_iR^{ind}_t+\epsilon^i_t$. Then sell to your clients $R^{ei}_t−b_iSP500_t−h_iETF_t−c_iR^{ind}_t=\alpha_i+\epsilon^i_t$. This is called "portable alpha."**

####Question 11
A buddy runs the CAPM regression. $R^i_{t+1}−R^f_t=\alpha_i+\beta_i,R^m[R^m_{t+1}−R^f_t]+\epsilon^i_{t+1}$. Which of the following could be, if true, a valid criticism of this regression? 

>**Answer: The CAPM is not about the time series regression. It is only about whether the alphas are zero. The CAPM is about the cross sectional relation between means and betas. That said, good tests have an economically and statistically siginificant spread in betas across their portfolios. If all the betas are not statistically different from one, they are likely not that different from each other. But that should be checked with a proper test on differences in betas. The main point was to disabuse you of the common fallacy that the quality of the time series regression is the measure of the CAPM.
The APT wants high $R^2$, but it is irrelevant for the CAPM. In fact, a major test of the CAPM is that only systematic risk is priced -- even stocks with large variance but little beta or correlation with the market should have low expected returns. These are precisely stocks with low $R^2$. So no, high $R^2$ is not that important.
Low $R^2$ does mean that test statistics will be imprecise; the betas will be poorly measured and confidence intervals will be large. That's one reason we usually use portfolios. But there is no theoretical reason to need high $R^2$. The CAPM says $\alpha=0$. Period.
The CAPM "time series" regression that defines beta,
$$R^i_t−R^f_t=\alpha_i+\beta_{i,m}(R^m_t−R^f_t)+\epsilon^i_t$$
has nothing whatsoever to do with the random walk regression
$$R^i_t−R^f_t=a_i+b_iX_{t−1}+\epsilon^i_t$$
The CAPM regression tells you about the correlation between the market today and the return today, and its expected return is about the differences across assets not over time. The random-walk regression is about whether something yesterday can predict a return today.
There are two uses of the word "factor model" -- "factor pricing" that expected returns depend on betas, with zero alphas here, and "factor models of variance" as in principal components. If the errors are correlated with each other, we know there are more variance factors. But that does not tell us the CAPM is not adequate for means. We know, for example, that stocks in the same industry move together. But those industry risks do not seem to generate higher mean returns. **
